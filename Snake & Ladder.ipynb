```python
import random
```

```python
def dice():
    return random.randint(1,6)
```

```python
# snakes position
def check_position(pos=0):
    snake={33:5,54:34,63:16,93:74,97:61}
    
    
    for x in snake:
        if pos==x:
            print("\n<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<")
            print("Oops! A snake : ",snake[x])
            print("<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<")
            pos=snake[x]
            break
        else:
            continue
    ladder={2:38,8:31,21:42,46:84,51:67,71:91,80:99}
    for y in ladder:
        if pos==y:
            print("\n>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
            print("Yay!! A ladder : ",ladder[y])
            print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
            pos=ladder[y]
            break
        else:
            continue
    return pos
```

```python
# player1
def player1(pos=0):
    while pos<1:
        input("\nPlayer 1: Press enter to throw the dice: ")
        throw=dice()
        print("Player 1 got: ",throw)
        if throw==6 or throw ==1:
            pos=1
            return pos
        else:
            return pos
        
    while pos>0 and pos<=100:
        input("\nPlayer 1: Press enter to throw the dice: ")
        throw=dice()
        print("Player 1 got: ",throw)    
        pos+=throw
        if pos>100:
            pos-=throw
            print("No chance to move")
        pos=check_position(pos)
        return pos
```

```python
# player 2
def player2(pos=0):
   
    while pos<1:
        input("\nPlayer 2: Press enter to throw the dice: ")
        throw=dice()
        print("player 2 got: ",throw)
        if throw==6 or throw == 1:
            pos=1
            return pos
        else:
            return pos
    
    while pos>0 and pos<=100:
        input("\nPlayer 2: Press enter to throw the dice: ")
        throw=dice()
        print("Player 2 got: ",throw)
        pos+=throw
        
        if pos>100:
            pos-=throw
            print("No chance to move")
        pos=check_position(pos)
        return pos
```

```python
# game starts
x=input("\nPress enter to start the game: ")
a=0;b=0

while True:
    p1=player1(a)
    a=p1
    print("Player 1 is at: ",p1)
    p2=player2(b)
    b=p2
    print("Player 2 is at: ",p2)
    
    if p1>=100:
        print("\nHip Hip Hurray!!! \nPlayer 1 won")
        break
    elif p2>=100:
        print("\nHip Hip Hurray!!! \nPlayer 2 won")
        break
    else:
        continue
```

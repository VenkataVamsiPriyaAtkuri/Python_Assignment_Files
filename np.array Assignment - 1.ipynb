```python
# Write a program to access all the elements greater than 30 and less than 80 and multiples of 5 from an array of shape 10,10 .Elements range from 1 to 100
e = np.arange(1,101).reshape(10,10)
s_e = []
for i in e:
    for s in i:
        if s>30 and s<80 and s%5==0:
            s_e.append(s)
print(e)

print("\nelements that are greater than 30,less than 80 and multiples of 5 are: ")
print(s_e)
```

```python
#Write a program to find common values between two arrays.
a1 = np.array([10, 20, 40, 60])
a2 = np.array([10, 30, 40, 50])
i = np.intersect1d(a1,a2)
print(a1)
print(a2)

print("\nCommon values: \n",i)
```
